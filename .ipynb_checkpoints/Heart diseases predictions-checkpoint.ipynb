## Domain Analysis

There are 14 columns in the dataset, where the patient_id column is a unique and random identifier. The remaining 13 features are described in the section below.

1   -  slope_of_peak_exercise_st_segment (type: int): the slope of the peak exercise ST segment, an electrocardiography read out        indicating quality of blood flow    to the heart

2   -  thal (type: categorical): results of thallium stress test measuring blood flow to the heart, with possible values normal,        fixed_defect, reversible_defect

3   -  resting_blood_pressure (type: int): resting blood pressure

4   -  chest_pain_type (type: int): chest pain type (4 values)

5   -  num_major_vessels (type: int): number of major vessels (0-3) colored by flourosopy

6   -  fasting_blood_sugar_gt_120_mg_per_dl (type: binary): fasting blood sugar > 120 mg/dl

7   -  resting_ekg_results (type: int): resting electrocardiographic results (values 0,1,2)

8   -  serum_cholesterol_mg_per_dl (type: int): serum cholestoral in mg/dl

9   -  oldpeak_eq_st_depression (type: float): oldpeak = ST depression induced by exercise relative to rest, a measure of              abnormality in electrocardiograms 

10  -  sex (type: binary): 0: female, 1: male

11  -  age (type: int): age in years

12  -  max_heart_rate_achieved (type: int): maximum heart rate achieved (beats per minute)

13  -  exercise_induced_angina (type: binary): exercise-induced chest pain (0: False, 1: True)

## Data Preprocessing

```python
cate_val=[]
cont_val=[]

for column in data.columns:
    if data[column].nunique()<=10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cate_val
```

```python
cont_val
```

```python
cate_val.remove('sex')
cate_val.remove('heart_disease_present')
data= pd.get_dummies(data,columns = cate_val,drop_first = True)
```

```python

data['slope_of_peak_exercise_st_segment_2'] = data['slope_of_peak_exercise_st_segment_2'].astype(int)
data['slope_of_peak_exercise_st_segment_3'] = data['slope_of_peak_exercise_st_segment_3'].astype(int)

data['thal_2'] = data['thal_2'].astype(int)
data['thal_3'] = data['thal_3'].astype(int)

data['fasting_blood_sugar_gt_120_mg_per_dl_1'] = data['fasting_blood_sugar_gt_120_mg_per_dl_1'].astype(int)

data['chest_pain_type_2'] = data['chest_pain_type_2'].astype(int)
data['chest_pain_type_3'] = data['chest_pain_type_3'].astype(int)
data['chest_pain_type_4'] = data['chest_pain_type_4'].astype(int)

data['num_major_vessels_1'] = data['num_major_vessels_1'].astype(int)
data['num_major_vessels_2'] = data['num_major_vessels_2'].astype(int)
data['num_major_vessels_3'] = data['num_major_vessels_3'].astype(int)

data['resting_ekg_results_1'] = data['resting_ekg_results_1'].astype(int)
data['resting_ekg_results_2'] = data['resting_ekg_results_2'].astype(int)

data['exercise_induced_angina_1'] = data['exercise_induced_angina_1'].astype(int)
```

```python
data.head(10)
```

```python
st = StandardScaler()
data[cont_val] = st.fit_transform(data[cont_val])
```

```python
data.head()
```

## Model Creation

```python
x = data.drop('heart_disease_present',axis = 1)
```

```python
y =data['heart_disease_present']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size= 0.2)
```

```python
y_test
```

```python
data.head()
```

## Logistic Regression

```python
log = LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
y_pred1 = log.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
print("Classification Report")
print(classification_report(y_test, y_pred1))
```

## SVC

```python
svm = svm.SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
y_pred2 = svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
print("Classification Report")
print(classification_report(y_test, y_pred2))
```

## KNeighborsClassifier

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred3=knn.predict(x_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
print("Classification Report")
print(classification_report(y_test, y_pred3))
```

## Decision TreeClassifier

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred4 = dt.predict(x_test)
```

```python
accuracy_score(y_test,y_pred4)
```

```python
print("Classification Report")
print(classification_report(y_test, y_pred4))
```

## Random forest

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred5 = rf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred5)
```

```python
print("Classification Report")
print(classification_report(y_test, y_pred5))
```
